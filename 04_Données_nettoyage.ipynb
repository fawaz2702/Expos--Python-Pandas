### 4.1.Cellules vides
Les cellules vides peuvent potentiellement vous donner un mauvais résultat lorsque vous analysez des données.

### 4.2.Supprimer des lignes
Une façon de gérer les cellules vides consiste à supprimer les lignes qui contiennent des cellules vides.

C'est généralement correct, car les ensembles de données peuvent être très volumineux et la suppression de quelques lignes n'aura pas un grand impact sur le résultat.

Exemple
Renvoie un nouveau bloc de données sans cellules vides :



```python
import pandas as pd

df = pd.read_csv('data.csv')

new_df = df.dropna()

print(new_df.to_string())
```

Remarque : par défaut, la dropna()méthode renvoie un nouveau DataFrame et ne modifie pas l'original.

Si vous souhaitez modifier le DataFrame d'origine, utilisez l' inplace = Trueargument :

Exemple
Supprimez toutes les lignes avec des valeurs NULL :

```python
import pandas as pd

df = pd.read_csv('data.csv')

df.dropna(inplace = True)

print(df.to_string())
```

Remarque : Maintenant, le dropna(inplace = True)ne renverra PAS de nouveau DataFrame, mais il supprimera toutes les lignes contenant des valeurs NULL du DataFrame d'origine.

### 4.3.Remplacer les valeurs vides
Une autre façon de gérer les cellules vides consiste à insérer une nouvelle valeur à la place.

De cette façon, vous n'avez pas à supprimer des lignes entières simplement à cause de certaines cellules vides.

La fillna()méthode nous permet de remplacer les cellules vides par une valeur :

Exemple
Remplacez les valeurs NULL par le nombre 130 :

```python
import pandas as pd

df = pd.read_csv('data.csv')

df.fillna(130, inplace = True)
print(df.to_string())
```

### 4.4.Remplacer uniquement pour une colonne spécifiée
L'exemple ci-dessus remplace toutes les cellules vides dans l'ensemble du bloc de données.

Pour remplacer uniquement les valeurs vides d'une colonne, spécifiez le nom de la colonne pour le DataFrame :

Exemple
Remplacez les valeurs NULL dans les colonnes "Calories" par le nombre 130 :

```python
import pandas as pd

df = pd.read_csv('data.csv')

df["Calories"].fillna(54452, inplace = True)
print(df.to_string())
```

### 4.5.Remplacer à l'aide de la moyenne, de la médiane ou du mode
Une façon courante de remplacer des cellules vides consiste à calculer la valeur moyenne, médiane ou de mode de la colonne.

Pandas utilise les méthodes mean() median()et mode()pour calculer les valeurs respectives d'une colonne spécifiée :

Exemple
Calculez la MOYENNE et remplacez toutes les valeurs vides par celle-ci :

```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)
print(df.to_string())
```

Médiane = la valeur au milieu, après avoir trié toutes les valeurs par ordre croissant.

Exemple
Calculez le MODE et remplacez les valeurs vides par celui-ci :



```python
import pandas as pd

df = pd.read_csv('data.csv')

x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace = True)
print(df.to_string())
```

Mode = la valeur qui apparaît le plus fréquemment.

## 4.6.Pandas - Correction de données erronées

### 4.6.1.Données erronées
Les « données erronées » ne doivent pas nécessairement être des « cellules vides » ou un « mauvais format », cela peut simplement être faux, comme si quelqu'un avait enregistré « 199 » au lieu de « 1.99 ».

Parfois, vous pouvez repérer des données erronées en examinant l'ensemble de données, car vous vous attendez à ce qu'elles devraient être.

Si vous jetez un œil à notre ensemble de données, vous pouvez voir qu'à la ligne 7, la durée est de 450, mais pour toutes les autres lignes, la durée est comprise entre 30 et 60.

Cela ne doit pas être faux, mais étant donné qu'il s'agit de l'ensemble de données des séances d'entraînement de quelqu'un, nous concluons avec le fait que cette personne ne s'est pas entraînée en 450 minutes.
Comment pouvons-nous corriger les valeurs erronées, comme celle de "Durée" à la ligne 7 ?

### 4.6.2.Remplacement des valeurs
Une façon de corriger les valeurs erronées est de les remplacer par autre chose.

Dans notre exemple, il s'agit très probablement d'une faute de frappe, et la valeur doit être "45" au lieu de "450", et nous pourrions simplement insérer "45" dans la ligne 7 :

Exemple
Définissez « Durée » = 45 à la ligne 7 :

```python
df.loc[7, 'Duration'] = 45
print(df.to_string())
```

Pour les petits ensembles de données, vous pourrez peut-être remplacer les mauvaises données une par une, mais pas pour les grands ensembles de données.

Pour remplacer des données erronées pour des ensembles de données plus volumineux, vous pouvez créer des règles, par exemple définir des limites pour les valeurs légales et remplacer toutes les valeurs qui se trouvent en dehors des limites.

Exemple
Parcourez toutes les valeurs de la colonne "Durée".

Si la valeur est supérieure à 120, réglez-la sur 120 :

```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 12025
print(df.to_string())   
```

### 4.6.3.Supprimer des lignes
Une autre façon de gérer des données erronées consiste à supprimer les lignes contenant des données erronées.

De cette façon, vous n'avez pas à trouver par quoi les remplacer, et il y a de fortes chances que vous n'en ayez pas besoin pour faire vos analyses.

Exemple
Supprimez les lignes où la "Durée" est supérieure à 120 :

```python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.drop(x, inplace = True)
```

## 4.7.Pandas - Suppression des doublons

Pour découvrir les doublons, nous pouvons utiliser la duplicated()méthode.

La duplicated()méthode renvoie une valeur booléenne pour chaque ligne :

Exemple
Renvoie Truepour chaque ligne qui est un doublon, sinon False:

```python
print(df.duplicated())
```

### 4.7.2.Suppression des doublons
Pour supprimer les doublons, utilisez la drop_duplicates()méthode.

Exemple
Supprimez tous les doublons :

```python
df.drop_duplicates(inplace = True)
print(df.duplicated())
```
